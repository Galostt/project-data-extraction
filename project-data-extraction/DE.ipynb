```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
forbes= requests.get('https://forbes.co/2021/04/09/editors-picks/los-50-mas-ricos-del-mundo-2021/')
soup = BeautifulSoup(forbes.content, "lxml")
#soup.select('figure.wp-block-image ~ p')
```

```python
top= [x.text for x in soup.select('h2')][1:41]
top2= [x.text for x in soup.select('figure.wp-block-image + p')][:40]
top3= [x.text for x in soup.select('figure.wp-block-image + * + p')][:40]
```

```python
list= {"nombre": top, "cantidad":top2, "nacionalidad":top3}
pdf= pd.DataFrame(data=list)
pdf
```
